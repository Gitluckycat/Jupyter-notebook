```python
import random
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import utils
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import pylab as mpl
import seaborn as sns

mpl.rcParams['font.sans-serif'] = ['SimHei']
mpl.rcParams['axes.unicode_minus'] = False

data = pd.read_csv(r"C:\Users\lenovo\Desktop\3.0.csv")
wm_data = utils.shuffle(data)


Y = wm_data["好瓜"].values
X = wm_data.drop("好瓜", axis=1).values

encoder = LabelEncoder()
y_label = encoder.fit_transform(Y)
color_label = LabelEncoder().fit_transform(wm_data['色泽'].values)
gen_shape = LabelEncoder().fit_transform(wm_data["根蒂"].values)
lisen_label = LabelEncoder().fit_transform(wm_data["敲声"].values)
wenli = LabelEncoder().fit_transform(wm_data['纹理'].values)
shape_label = LabelEncoder().fit_transform(wm_data["脐部"].values)
chugan_label = LabelEncoder().fit_transform(wm_data["触感"].values)

feature_list = list(wm_data.columns[1:len(wm_data.columns)-1])
target_names = wm_data.columns[len(wm_data.columns)-1]

print(feature_list)
print(wm_data.head())
print(wm_data.describe())


def plot_scatter(wm_data):
    plt.figure(figsize=(16, 10))
    for i in range(6):
        for j in range(6):
            plt.subplot(6, 6, i*6+j+1)
            if i == j:
                plt.hist(X[:, i+1], color='gray')
                plt.xlabel(feature_list[i])

            else:
                plt.scatter(X[:, j+1], X[:, i+1], c=y_label, cmap='brg')
                plt.xlabel(feature_list[j])
                plt.ylabel(feature_list[i])
    plt.suptitle('Pairwise scatter plot of watermelon features')
    plt.tight_layout()
    plt.show()


plot_scatter(wm_data)

# wm_data.drop(['编号'], axis=1, inplace=True)
std_1 = MinMaxScaler()
std_2 = MinMaxScaler()
std_3 = MinMaxScaler()
std_4 = MinMaxScaler()
std_5 = MinMaxScaler()
std_6 = MinMaxScaler()
std_7 = MinMaxScaler()
wm_data['好瓜'] = std_1.fit_transform(y_label.reshape(-1, 1))
wm_data["色泽"] = std_2.fit_transform(color_label.reshape(-1, 1))
wm_data['根蒂'] = std_3.fit_transform(gen_shape.reshape(-1, 1))
wm_data['敲声'] = std_4.fit_transform(lisen_label.reshape(-1, 1))
wm_data['纹理'] = std_5.fit_transform(wenli.reshape(-1, 1))
wm_data['脐部'] = std_6.fit_transform(shape_label.reshape(-1, 1))
wm_data['触感'] = std_7.fit_transform(chugan_label.reshape(-1, 1))

plt.figure(figsize=(10, 8))
corr = wm_data.corr()
print(corr)
sns.heatmap(corr, annot=True, cmap='RdBu')
plt.show()


wm_data.plot(x='好瓜', kind='bar')
plt.show()







```
