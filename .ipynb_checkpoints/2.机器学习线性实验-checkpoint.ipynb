## 内容1：利用梯度下降法求解$y=x^2$在区间[-10,10]的最小值
梯度下降算法的基本思想是通过迭代优化，通过不断地调整参数值，使得损失函数逐渐减小，从而找到最小值点。这个过程中，算法利用梯度的信息指引调整方向和步长，从而更快地接近最小值。学习率的选择非常重要，过大的学习率可能导致算法无法收敛，而过小的学习率可能导致算法收敛速度过慢。

梯度下降法的具体步骤如下：
1. 初始化参数：选择一个初始参数值作为起始点。
1. 计算损失函数梯度：计算当前参数值的梯度，梯度表示了函数在该点上升最快的方向。
1. 更新参数：通过沿着负梯度方向减小参数值，来使损失函数逐渐减小。参数更新的步长由学习率决定。
1. 判断终止条件：检查算法是否已经收敛，也就是损失函数是否已经足够小或者参数是否已经稳定不变。
1. 如果未达到终止条件，则返回第2步继续执行。否则，返回最终参数。

```python
import numpy as np
import maplotlib.pyplot as plt
# 定义函数f(x)=x^2
def f(x):
    return np.power(x,2)

# 定义函数求导公式
def df(x):
    return 2.0*x
```

```python
# 限制自变量x的范围
import matplotlib.pyplot as plt
xs = np.arange(-10,11)
# 绘图，查看原函数图
plt.plot(xs,f(xs))
plt.show()
```

```python
# 学习率（步长）
learning_rate = 0.1
# 最大迭代次数
max_loop = 50
# 最小梯度值退出
min_delta = 0.001
# x初始值
x_init = 10.0
x = x_init
```

```python
# 循环进行梯度下降
dict_Loop_X = {}
dict_Loop_X[0] = x
for i in range(max_loop):
    df_x = df(x)
    if(df_x <= min_delta):
        break
    x = x - learning_rate * df_x
    # list_X.append(x)
    dict_Loop_X[i] = x
    print(x)
```

```python
# 查看梯度下降优化后的情况
print("init x =", x_init)
print("arg min f(x) of x = ", x)
print("f(x)=",f(x))
```

```python

np_X = np.array(list(dict_Loop_X.keys()))
np_Value_Y = np.array(list(dict_Loop_X.values()))

plt.plot(np_X,np_Value_Y)
plt.show()
```

```python
# 展示梯度下降迭代次数和f(X)之间的关系，可以看随着迭代次数增加，f(X)越来越接近0
np_X = np.array(list(dict_Loop_X.keys()))
np_Value_Y = np.array(list(dict_Loop_X.values()))

plt.plot(np_X,f(np_Value_Y))
plt.show()
```

编程实现对数几率回归，利用梯度下降法进行训练，并给出西瓜数据集3.0a上的结果。

提示：

实验要求：

```python
# 导入必要的库
import numpy as np
import pandas as pd

# 读取西瓜数据集3.0
data = pd.read_csv(r"C:\Users\lenovo\Desktop\3.0.csv", usecols=["色泽","根蒂","敲声","纹理","脐部","触感","密度", "含糖率", "好瓜"], encoding="gbk")
data.columns = ['色泽', '根蒂', '敲声', '纹理', '脐部', '触感', '密度', '含糖率', '好瓜']

# 将离散属性转换为数值型
data['色泽'] = data['色泽'].map({'青绿': 0, '乌黑': 1, '浅白': 2})
data['根蒂'] = data['根蒂'].map({'蜷缩': 0, '稍蜷': 1, '硬挺': 2})
data['敲声'] = data['敲声'].map({'浊响': 0, '沉闷': 1, '清脆': 2})
data['纹理'] = data['纹理'].map({'清晰': 0, '稍糊': 1, '模糊': 2})
data['脐部'] = data['脐部'].map({'凹陷': 0, '稍凹': 1, '平坦': 2})
data['触感'] = data['触感'].map({'硬滑': 0, '软粘': 1})

# 将好瓜的标签编码为1，坏瓜的标签编码为0
data['好瓜'] = data['好瓜'].map({'是': 1, '否': 0})

# 将数据分为特征和标签
X = data.iloc[:, :-1].values # 特征
y = data.iloc[:, -1].values # 标签

# 对数据进行归一化处理
X_min = np.min(X, axis=0) # 特征的最小值
X_max = np.max(X, axis=0) # 特征的最大值
X = (X - X_min) / (X_max - X_min) # 归一化后的特征

# 定义对数回归模型
def logistic_regression(X, y, alpha=0.01, max_iter=4000):
    # 初始化权重
    w = np.zeros(X.shape[1])
    # 迭代更新权重
    for i in range(max_iter):
        # 计算预测概率
        z = np.dot(X, w)
        y_pred = 1 / (1 + np.exp(-z))
        # 计算损失函数
        loss = -np.mean(y * np.log(y_pred) + (1 - y) * np.log(1 - y_pred))
        # 计算梯度
        grad = np.dot(X.T, (y_pred - y)) / X.shape[0]
        # 更新权重
        w = w - alpha * grad
        # 打印损失函数
        if(i%50==0):
            print(f'Iteration {i+1}, loss: {loss}')
    # 返回权重
    return w

# 训练模型
w = logistic_regression(X, y)

# 预测测试数据
y_pred = 1 / (1 + np.exp(-np.dot(X, w)))
y_pred = np.where(y_pred >= 0.5, 1, 0)

# 计算准确率
accuracy = np.mean(y_pred == y)
print(f'Accuracy: {accuracy}')
```

```python
%reset -f
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
random_state = 60090

# 使用make_classification生成分类样本数据，利用random_state指定随机种子，便于生成相同的分类样本数据，n_samples和n_features等参数未指定，使用默认数值。生成的X为特征矩阵，生成的y为标签向量
X, y = make_classification(random_state=random_state)
# train_test_split划分数据集为训练集和验证集，这两个集中又包括特征矩阵和标签向量
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=random_state)
# 利用skleran库中的LogisticRegression来创建逻辑回归对象clf
clf = LogisticRegression(penalty=None, )
# fit函数是将x_train训练数据，y_train是标签数据用clf对象表示的逻辑回归函数进行预训练
clf.fit(X_train, y_train)
# 利用clf对象自带的预测函数predict来预测在验证数据为x_test的情况下，得到该模型预测的标签数据
y_pred = clf.predict(X_test)
# 利用accuracy_score输出预测的准确率
print("Accuracy:", accuracy_score(y_test, y_pred))
# 使用逻辑回归的内置属性返回模型的参数，coef_和intercept_都是模型的参数，intercept_返回的是偏置项，而coef_返回的是线性表达式中的w
print("Coefficients:", clf.coef_, clf.intercept_)
```

```python
%reset -f
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split, KFold, LeaveOneOut, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, precision_recall_curve, auc, roc_curve

random_state = 60090

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams["figure.figsize"] = (5, 2)

X, y = make_classification(random_state=random_state)

X_min = np.min(X, axis=0) # 特征的最小值
X_max = np.max(X, axis=0) # 特征的最大值
X = (X - X_min) / (X_max - X_min) # 归一化后的特征
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=random_state)




def linear(w, b, X):
    # z = ?
    z = np.dot(X, w) + b
    return z

def sigmoid(z):
    # y_hat = ?
    y_hat = 1/(1+np.exp(-z))
    return y_hat

def log_likelihood_loss(y_hat, y):
    # loss = ?
    loss = -np.mean(y * np.log(y_hat) + (1 - y) * np.log(1 - y_hat))
    return loss

def gradient_descent(w, b, X, y, eta):
    # w, b = ?
    b_init = np.array([1 for _ in range(X.shape[0])])
    grad_w = np.dot(X.T, (sigmoid(linear(w, b, X)) - y)) / X.shape[0]
    grad_b = np.dot(b_init.T, (sigmoid(linear(w, b, X)) - y)) / X.shape[0] 
    w = w - eta*grad_w
    b = b - eta*grad_b
    return w, b

class LogisticRegression_own(object):
    def __init__(self, eta=0.01):
        self.w = None
        self.b = None
        self.eta = eta

    def fit(self, X, y, iterations = 10000):
        self.w = np.random.randn(X.shape[-1])
        self.b = np.random.randn()
        for cnt in range(iterations):
            self.w, self.b = gradient_descent(self.w, self.b, X, y, self.eta)
            if cnt % 1000 == 0:
                print(f'Iteration {cnt}, loss: {log_likelihood_loss(sigmoid(linear(self.w, self.b, X)), y)}')

    def predict(self, X):
        y_test_pred = sigmoid(linear(self.w, self.b, X))
        return np.where(y_test_pred > 0.5, 1, 0)

    def predict_proba(self, X):
        y_test_pred = sigmoid(linear(self.w, self.b, X))
        return y_test_pred
        
        
clf_2 = LogisticRegression_own()
clf_2.fit(X_train, y_train)
y_pred = clf_2.predict(X_test)
y_pred_proba_1 = clf_2.predict_proba(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print(clf_2.w, clf_2.b)

```

```python
def ruc_curve(y_true, y_pred):
    fpr, tpr, threshold = roc_curve(y_true, y_pred)
    rocauc = auc(fpr, tpr)
    print("AUC的值是：", rocauc)
    plt.plot(fpr, tpr, 'b', label='AUC = %0.2f' % rocauc)
    plt.legend(loc='lower right')
    plt.plot([0, 1], [0, 1], 'r--')
    plt.title(f'自制逻辑回归模型的AUC值是：{rocauc}')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.ylabel('true rate')
    plt.xlabel('false rate')
    plt.show()

    
ruc_curve(y_test, y_pred)
```

## 内容6：在理解内容3和5的基础上，利用sklearn编写逻辑回归算法，实现对西瓜数据集3.0预测

```python
# 此处完成内容6的代码实现
import random
import numpy as np
import pandas as pd
from sklearn import utils
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
number = 60090

plt.rcParams['axes.unicode_minus']=False
plt.rcParams['font.sans-serif'] = ['SimHei']

wm_data = pd.read_csv(r"C:\Users\lenovo\Desktop\3.0.csv")
wm_data['色泽'] = wm_data['色泽'].map({'青绿': 0, '乌黑': 1, '浅白': 2})
wm_data['根蒂'] = wm_data['根蒂'].map({'蜷缩': 0, '稍蜷': 1, '硬挺': 2})
wm_data['敲声'] = wm_data['敲声'].map({'浊响': 0, '沉闷': 1, '清脆': 2})
wm_data['纹理'] = wm_data['纹理'].map({'清晰': 0, '稍糊': 1, '模糊': 2})
wm_data['脐部'] = wm_data['脐部'].map({'凹陷': 0, '稍凹': 1, '平坦': 2})
wm_data['触感'] = wm_data['触感'].map({'硬滑': 0, '软粘': 1})
wm_data['好瓜'] = wm_data['好瓜'].map({'是': 1, '否': 0})
y = wm_data["好瓜"].values
X = wm_data.drop("好瓜", axis=1).values

X_min = np.min(X, axis=0) # 特征的最小值
X_max = np.max(X, axis=0) # 特征的最大值
X = (X - X_min) / (X_max - X_min) # 归一化后的特征

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2 ,random_state = number)
theta = np.ones(X.shape[-1]-1)

class logisticRegression():
    def __init__(self, eta=0.01):
        self.stop_iter = 0
        self.stop_cost = 1
        self.stop_grad = 2
        self.eta = eta
        

    def fit(self, X, y, stop_type, theta, batchsize, threshold, iterations=10000):
        grad = np.zeros(theta.shape)
        costs = [self.cost(X[:, 1:], y, theta)]
        iter = 0
        
        while iter <= iterations:
            train_list = []
                
            for i in range(batchsize):
                 train_list.append(random.randint(0, X.shape[0] -1))
            
            grad_iter = 0
            for i in range(batchsize):
                grad = self.gradient(X[train_list[i], 1:], y[train_list[i]], theta)
                grad_iter += grad
                
            grad_iter = grad_iter/batchsize
            theta = theta - grad_iter*self.eta
           
            iter += 1
            costs.append(self.cost(X[:, 1:], y, theta))
            

            if stop_type == self.stop_iter:
                 value = iter
            elif stop_type == self.stop_cost:
                 value = costs
            elif stop_type == self.stop_grad:
                 value = grad
            if self.stop_citerion(stop_type, value, threshold):
                 break

        return theta, grad, iter-1, costs

    def sigmoid(self, z):
        y_pre = 1/(1+np.exp(-z))
        return y_pre

    def model(self, X, theta):
        z = np.dot(X, theta)
        return self.sigmoid(z)

    
    def gradient(self, X, y, theta):
        err = (self.model(X, theta) - y)
        grad_theta = np.dot(X.T, err) 

        return grad_theta

    def cost(self, X, y, theta):
        loss = -np.mean(y*np.log(self.model(X, theta)) + (1 - y)*np.log(1-self.model(X, theta)))
        return loss
        

    def stop_citerion(self, stop_type, value, threshold):
        if stop_type == self.stop_iter:
            return value > threshold
        elif stop_type == self.stop_cost:
            return abs(value[-1]-value[-2]) < threshold
        elif stop_type == self.stop_grad:
            return np.linalg.norm(value) < threshold

    def predict(self, X, theta):
        y_pred = self.model(X[:, 1:], theta)
        return np.where(y_pred>0.5, 1, 0)
```

```python
clf = logisticRegression()
theta, grad, itera, costs = clf.fit(x_train, y_train, 0, theta, 3, 20000)
y_pred = clf.predict(x_test, theta)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
def runplot(x_train, y_train, theta, batchSize, stop_type, thresh, alpha):
    clf.eta = alpha
    tit = '停止条件为：'
    theta, grad, iter, costs= clf.fit(x_train, y_train, stop_type, theta, batchSize, thresh)
    # plt.subplots(figsize=(12,4))
    if stop_type == clf.stop_iter: 
        strStop = f"{thresh} iterations"
    elif stop_type == clf.stop_cost: 
        strStop = f"costs change < {thresh}"
    else: 
        strStop = f"gradient norm < {thresh}"
    tit = tit + strStop
    tit = tit + '的逻辑回归模型'
    plt.plot(np.arange(len(costs)), costs, 'r')
    plt.title(tit)
    plt.xlabel('Iterations')
    plt.ylabel('Cost') 
    plt.xlim(-1,)
    return theta

```

```python
%reset -f
import pandas as pd
import numpy as np
import matplotlib.pyplot
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split, KFold, LeaveOneOut
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
random_state = 60090

wm_data = pd.read_csv(r"C:\Users\lenovo\Desktop\3.0.csv")
wm_data['色泽'] = wm_data['色泽'].map({'青绿': 0, '乌黑': 1, '浅白': 2})
wm_data['根蒂'] = wm_data['根蒂'].map({'蜷缩': 0, '稍蜷': 1, '硬挺': 2})
wm_data['敲声'] = wm_data['敲声'].map({'浊响': 0, '沉闷': 1, '清脆': 2})
wm_data['纹理'] = wm_data['纹理'].map({'清晰': 0, '稍糊': 1, '模糊': 2})
wm_data['脐部'] = wm_data['脐部'].map({'凹陷': 0, '稍凹': 1, '平坦': 2})
wm_data['触感'] = wm_data['触感'].map({'硬滑': 0, '软粘': 1})
wm_data['好瓜'] = wm_data['好瓜'].map({'是': 1, '否': 0})
y = wm_data["好瓜"].values
X = wm_data.drop("好瓜", axis=1).values

X_min = np.min(X, axis=0) # 特征的最小值
X_max = np.max(X, axis=0) # 特征的最大值
X = (X - X_min) / (X_max - X_min) # 归一化后的特征
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=random_state)


def linear(w, b, X):
    # z = ?
    z = np.dot(X, w) + b
    return z

def sigmoid(z):
    # y_hat = ?
    y_hat = 1/(1+np.exp(-z))
    return y_hat

def log_likelihood_loss(y_hat, y):
    # loss = ?
    loss = -np.mean(y * np.log(y_hat) + (1 - y) * np.log(1 - y_hat))
    return loss

def gradient_descent(w, b, X, y, eta):
    # w, b = ?
    b_init = np.array([1 for _ in range(X.shape[0])])
    grad_w = np.dot(X.T, (sigmoid(linear(w, b, X)) - y)) / X.shape[0]
    grad_b = np.dot(b_init.T, (sigmoid(linear(w, b, X)) - y)) / X.shape[0] 
    w = w - eta*grad_w
    b = b - eta*grad_b
    return w, b

class LogisticRegression_own(object):
    def __init__(self, eta=0.01):
        self.w = None
        self.b = None
        self.eta = eta

    def fit(self, X, y, iterations = 10000):
        self.w = np.random.randn(X.shape[-1])
        self.b = np.random.randn()
        for cnt in range(iterations):
            self.w, self.b = gradient_descent(self.w, self.b, X, y, self.eta)
            if cnt % 1000 == 0:
                print(f'Iteration {cnt}, loss: {log_likelihood_loss(sigmoid(linear(self.w, self.b, X)), y)}')

    def predict(self, X):
        y_test_pred = sigmoid(linear(self.w, self.b, X))
        return np.where(y_test_pred > 0.5, 1, 0)

clf = LogisticRegression()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print(clf.coef_, clf.intercept_)


```
